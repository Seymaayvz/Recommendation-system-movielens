```python
import pandas as pd
import numpy as np

from sklearn.metrics.pairwise import cosine_similarity
```

```python
cols = ["user_id", "movie_id", "rating", "timestamp"]

df = pd.read_csv(
    "../data/raw/u.data",
    sep="\t",
    names=cols
)

df.head()
```

```python
user_movie_matrix = df.pivot_table(
    index="user_id",
    columns="movie_id",
    values="rating"
)

user_movie_matrix.head()
```

```python
user_movie_matrix_filled = user_movie_matrix.fillna(0)
```

```python
item_similarity = cosine_similarity(
    user_movie_matrix_filled.T
)

item_similarity_df = pd.DataFrame(
    item_similarity,
    index=user_movie_matrix_filled.columns,
    columns=user_movie_matrix_filled.columns
)

item_similarity_df.head()
```

```python
def recommend_movies(movie_id, top_n=5):
    similar_movies = (
        item_similarity_df[movie_id]
        .sort_values(ascending=False)
        .iloc[1:top_n+1]
    )
    return similar_movies
```

```python
recommend_movies(movie_id=1, top_n=5)
```

```python
# User-Item Matrix

user_item_matrix = df.pivot_table(
    index="user_id",
    columns="movie_id",
    values="rating"
)

user_item_matrix.head()

# Bu matriste NaN değerler vardır çünkü her kullanıcı her filmi puanlamamıştır.
# Collaborative Filtering için bu durum normaldir.
```

```python
from sklearn.metrics.pairwise import cosine_similarity
user_similarity = cosine_similarity(user_item_matrix.fillna(0))
user_similarity_df = pd.DataFrame(
    user_similarity,
    index=user_item_matrix.index,
    columns=user_item_matrix.index
)

user_similarity_df.head()

```

```python
def get_similar_users(user_id, n=5):
    return (
        user_similarity_df[user_id]
        .sort_values(ascending=False)
        .iloc[1:n+1]
    )

get_similar_users(1)
```

```python
def recommend_movies(user_id, n=5):
    similar_users = get_similar_users(user_id).index

    user_ratings = user_item_matrix.loc[user_id]
    unseen_movies = user_ratings[user_ratings.isna()].index

    recommendations = (
        user_item_matrix.loc[similar_users, unseen_movies]
        .mean()
        .sort_values(ascending=False)
        .head(n)
    )

    return recommendations
```

```python
recommend_movies(1)
```

```python
# Bu çalışmada User-Based Collaborative Filtering yöntemi kullanılmıştır.
# Kullanıcı benzerliği cosine similarity ile hesaplanmıştır.
# Model, kullanıcının daha önce puanlamadığı filmleri önermektedir.
```
